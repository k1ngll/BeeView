Um diferencial desta an√°lise ser√° a inclus√£o de aspectos qualitativos complexos, como:

## 5. Primeiras Observa√ß√µes e Direcionamentos da An√°lise

Pesquisas iniciais j√° apontam para tend√™ncias relevantes que ser√£o exploradas com os dados:

* **Desempenho de Sequ√™ncias vs. Originais:**
    * Estudos indicam que, em m√©dia, os filmes originais tendem a ter uma bilheteria e avalia√ß√£o ligeiramente superiores √†s suas sequ√™ncias diretas.
    * Contudo, sequ√™ncias ainda podem ser altamente lucrativas e, em alguns casos, superar os originais, especialmente em franquias bem estabelecidas.

* **A Relev√¢ncia da Avalia√ß√£o do P√∫blico:**
    * A recep√ß√£o cr√≠tica e a avalia√ß√£o do p√∫blico do filme original t√™m um impacto direto no interesse e no desempenho inicial de uma sequ√™ncia.
    * Uma avalia√ß√£o negativa pode comprometer significativamente o potencial de uma continua√ß√£o.

* **O Fator Tempo:**
    * O intervalo entre os lan√ßamentos √© crucial. Um per√≠odo √≥timo parece situar-se entre 2 a 5 anos.
    * Intervalos muito curtos podem saturar o p√∫blico, enquanto esperas muito longas podem diminuir o interesse e o "momentum" da franquia.

* **G√™neros com Maior Propens√£o a Franquias de Sucesso:**

* **Impacto de Mudan√ßas Dr√°sticas:**
    * **Morte de Atores Principais:** Pode levar a reescritas significativas, reescala√ß√µes (frequentemente com resist√™ncia dos f√£s), uso de CGI para recriar o ator ou a aposentadoria do personagem. O legado do ator e do personagem torna-se um ponto sens√≠vel para a continua√ß√£o.
    * **Sa√≠da de Diretores:** √â geralmente vista como um risco, podendo levar a uma queda na qualidade ou a uma mudan√ßa de tom que desagrade os f√£s. No entanto, novos diretores tamb√©m podem revitalizar uma franquia com novas perspectivas. A coes√£o da vis√£o criativa √© um desafio.

* **A Armadilha da Conclus√£o Narrativa:**
    * Continuar uma hist√≥ria que teve um desfecho satisfat√≥rio e completo exige uma justificativa narrativa muito forte.
    * Sequ√™ncias que parecem for√ßadas ou que desfazem o impacto emocional do final anterior tendem a ser mal recebidas.

#### 3.4 CRIANDO A TABELA ROI PARA A ANALISE FUTURA


```python
import numpy as np
import pandas as pd

# Certifique-se de que tmdb_dataset_filtrado est√° definido e com a coluna 'ROI' calculada.
# Se voc√™ n√£o executou o bloco de c√≥digo completo (carregamento + limpeza + c√°lculo de ROI)
# desde que iniciou ou resetou a sess√£o do Colab, por favor, execute-o primeiro.

# 1. Encontrar o valor m√°ximo de ROI que N√ÉO √© infinito
# Filtramos os valores infinitos para pegar o m√°ximo entre os valores finitos.
max_finite_roi = tmdb_dataset_filtrado[tmdb_dataset_filtrado['ROI'] != np.inf]['ROI'].max()

# Definir um valor de substitui√ß√£o para 'inf'.
# Usaremos 1.5 vezes o maior ROI finito. Isso garante que ele continua sendo o valor mais alto,
# mas agora √© um n√∫mero finito. Se por algum motivo n√£o houver ROI finito positivo, usamos um valor padr√£o alto.
if pd.isna(max_finite_roi) or max_finite_roi <= 0: # Caso n√£o haja ROI finito positivo ou seja 0
    replacement_value_for_inf = 1e9 # 1 bilh√£o, um n√∫mero arbitrariamente grande
else:
    replacement_value_for_inf = max_finite_roi * 1.5

# 2. Substituir os valores infinitos na coluna 'ROI'
# Substitu√≠mos tanto np.inf (infinito positivo) quanto -np.inf (infinito negativo, menos comum aqui)
tmdb_dataset_filtrado['ROI'].replace([np.inf, -np.inf], replacement_value_for_inf, inplace=True)

print(f"Valores 'inf' (e -inf) na coluna 'ROI' foram substitu√≠dos por: {replacement_value_for_inf:.2f}")

print("\n--- Estat√≠sticas Descritivas para a coluna ROI ap√≥s tratamento ---")
print(tmdb_dataset_filtrado['ROI'].describe())

print("\n--- Primeiras 5 linhas do DataFrame com o ROI tratado (observe o novo m√°ximo se houver) ---")
print(tmdb_dataset_filtrado.head())
```

```python
import pandas as pd
import numpy as np
import re # M√≥dulo 're' para express√µes regulares

# --- PASSO 0: CARREGAR O SEU DATASET DO KAGGLE ---
# Caminho do seu arquivo CSV no Google Drive
try:
    tmdb_dataset = pd.read_csv('/content/drive/MyDrive/Faculdade/Python BIG DATA/Zuum Filmes/TMDB_movie_dataset_25-05.csv')
    print("Dataset real carregado com sucesso do Google Drive!")
except FileNotFoundError:
    print("ERRO CR√çTICO: Arquivo do dataset n√£o encontrado no caminho especificado.")
    print("Por favor, verifique se o caminho est√° correto e se o Google Drive est√° montado no Colab.")
    raise # Levanta um erro para parar a execu√ß√£o, pois o dataset real √© essencial.


# --- PASSO 1: FILTRAR AS COLUNAS DESEJADAS ---
colunas_desejadas = [
    'id', 'title', 'vote_average', 'vote_count', 'revenue', 'budget',
    'popularity', 'genres', 'release_date', 'imdb_id',
    'production_countries', 'original_language'
]
tmdb_dataset_filtrado = tmdb_dataset[colunas_desejadas].copy()

print("\n--- Status do DataFrame antes da limpeza ---")
print(f"Total de linhas iniciais: {len(tmdb_dataset_filtrado)}")
print(tmdb_dataset_filtrado.info())
print("\n")


# --- PASSO 2: APLICA√á√ÉO DAS REGRAS DE LIMPEZA ---

# 1. Remover filmes onde 'budget' √© 0 E 'revenue' √© 0
linhas_para_remover_br = (tmdb_dataset_filtrado['budget'] == 0) & (tmdb_dataset_filtrado['revenue'] == 0)
tmdb_dataset_filtrado = tmdb_dataset_filtrado[~linhas_para_remover_br].copy()
print(f"Linhas removidas por 'budget' e 'revenue' zero: {linhas_para_remover_br.sum()}")
print(f"Total de linhas ap√≥s remover budget/revenue zero: {len(tmdb_dataset_filtrado)}")


# 2. Remover filmes com 'vote_count' igual a 0 ou menor que 50
limite_votos = 50
linhas_para_remover_votos = tmdb_dataset_filtrado['vote_count'] < limite_votos
tmdb_dataset_filtrado = tmdb_dataset_filtrado[~linhas_para_remover_votos].copy()
print(f"Linhas removidas por 'vote_count' < {limite_votos}: {linhas_para_remover_votos.sum()}")
print(f"Total de linhas ap√≥s remover filmes com poucos votos: {len(tmdb_dataset_filtrado)}")


# 3. Preencher 'genres' com 'Desconhecido'
tmdb_dataset_filtrado['genres'] = tmdb_dataset_filtrado['genres'].fillna('Desconhecido')
print("Valores nulos em 'genres' preenchidos com 'Desconhecido'.")


# 4. Converter 'release_date' para datetime e tratar nulos/inv√°lidos
tmdb_dataset_filtrado['release_date'] = pd.to_datetime(tmdb_dataset_filtrado['release_date'], errors='coerce')
print("Coluna 'release_date' convertida para datetime. Datas inv√°lidas/nulas s√£o agora NaT.")


# 5. Preencher 'imdb_id' com 'Desconhecido'
tmdb_dataset_filtrado['imdb_id'] = tmdb_dataset_filtrado['imdb_id'].fillna('Desconhecido')
print("Valores nulos em 'imdb_id' preenchidos com 'Desconhecido'.")


# 6. Preencher 'production_countries' com 'Desconhecido'
tmdb_dataset_filtrado['production_countries'] = tmdb_dataset_filtrado['production_countries'].fillna('Desconhecido')
print("Valores nulos em 'production_countries' preenchidos com 'Desconhecido'.")


# 7. Preencher 'title' com 'T√≠tulo Desconhecido'
tmdb_dataset_filtrado['title'] = tmdb_dataset_filtrado['title'].fillna('T√≠tulo Desconhecido')
print("Valores nulos em 'title' preenchidos com 'T√≠tulo Desconhecido'.")


print("\n--- Status do DataFrame ap√≥s a limpeza ---")
print(f"Total de linhas finais: {len(tmdb_dataset_filtrado)}")
print(tmdb_dataset_filtrado.info())
print("\n--- Estat√≠sticas Descritivas ap√≥s a limpeza ---")
print(tmdb_dataset_filtrado[['budget', 'revenue', 'vote_average', 'popularity', 'vote_count']].describe())


# --- PASSO 3: CALCULAR O ROI ---
tmdb_dataset_filtrado['ROI'] = (tmdb_dataset_filtrado['revenue'] - tmdb_dataset_filtrado['budget']) / tmdb_dataset_filtrado['budget']

# Tratamento do ROI Infinito
max_finite_roi = tmdb_dataset_filtrado[tmdb_dataset_filtrado['ROI'] != np.inf]['ROI'].max()
if pd.isna(max_finite_roi) or max_finite_roi <= 0:
    replacement_value_for_inf = 1e9
else:
    replacement_value_for_inf = max_finite_roi * 1.5
tmdb_dataset_filtrado['ROI'].replace([np.inf, -np.inf], replacement_value_for_inf, inplace=True)

print(f"Valores 'inf' (e -inf) na coluna 'ROI' foram substitu√≠dos por: {replacement_value_for_inf:.2f}")
print("\n--- Estat√≠sticas Descritivas para a coluna ROI ap√≥s tratamento ---")
print(tmdb_dataset_filtrado['ROI'].describe())

# --- AGORA, O C√ìDIGO PARA VERIFICAR PA√çSES E IDIOMAS NOVAMENTE ---
print("\n--- Idiomas Originais Dispon√≠veis (original_language) ---")
idiomas_disponiveis = tmdb_dataset_filtrado['original_language'].unique().tolist()
if len(idiomas_disponiveis) > 50:
    print(f"Total de idiomas √∫nicos: {len(idiomas_disponiveis)}")
    print(idiomas_disponiveis[:50])
    print("...")
else:
    print(idiomas_disponiveis)

print("\n--- Pa√≠ses de Produ√ß√£o Dispon√≠veis (production_countries) ---")
paises_brutos_disponiveis = tmdb_dataset_filtrado['production_countries'].unique().tolist()
if len(paises_brutos_disponiveis) > 50:
    print(f"Total de combina√ß√µes/pa√≠ses brutos √∫nicos: {len(paises_brutos_disponiveis)}")
    print(paises_brutos_disponiveis[:50])
    print("...")
else:
    print(paises_brutos_disponiveis)

todos_os_paises_individuais = set()
for paises_str in tmdb_dataset_filtrado['production_countries'].unique():
    if isinstance(paises_str, str):
        if paises_str == 'Desconhecido' and len(tmdb_dataset_filtrado['production_countries'].unique()) > 1:
            continue
        paises_lista = [p.strip() for p in re.split(r'\||, |; ', paises_str)]
        todos_os_paises_individuais.update(paises_lista)

paises_individuais_ordenados = sorted(list(todos_os_paises_individuais))

print(f"\n--- Pa√≠ses de Produ√ß√£o Individuais Dispon√≠veis (production_countries desmembrado) ---")
if len(paises_individuais_ordenados) > 50:
    print(f"Total de pa√≠ses individuais √∫nicos: {len(paises_individuais_ordenados)}")
    print(paises_individuais_ordenados[:50])
    print("...")
else:
    print(paises_individuais_ordenados)
```

#### API trazendo os filmes brasileiros


```python
import requests
import time # Para adicionar um pequeno atraso entre as requisi√ß√µes √† API

# --- CONFIGURA√á√ÉO DA API TMDB ---
API_KEY = '1b5d8e773c2596486bd43142c6485e8c' # <<<< SUBSTITUA PELA SUA CHAVE REAL DO TMDB >>>>
BASE_URL = 'https://api.themoviedb.org/3/movie/'

# Dicion√°rio para armazenar informa√ß√µes das cole√ß√µes/franquias encontradas
# A chave ser√° o ID da cole√ß√£o, e o valor ser√° um dicion√°rio com nome e lista de filmes.
collections_found = {}
# Lista para armazenar informa√ß√µes dos filmes brasileiros que s√£o sequ√™ncias
brazilian_franchise_movies = []

print("Iniciando a busca por franquias para os filmes brasileiros...")
print(f"Processando {len(df_brasileiros)} filmes...")

# Iterar sobre cada filme no DataFrame de filmes brasileiros
for index, row in df_brasileiros.iterrows():
    movie_id = row['id']
    movie_title = row['title']

    # Construir a URL da requisi√ß√£o para detalhes do filme
    url = f"{BASE_URL}{movie_id}?api_key={API_KEY}&language=pt-BR" # pt-BR para obter nomes de cole√ß√µes em portugu√™s se dispon√≠veis

    try:
        response = requests.get(url)
        response.raise_for_status() # Levanta um erro para c√≥digos de status HTTP ruins (4xx ou 5xx)
        movie_data = response.json()

        # Verificar se o filme pertence a uma cole√ß√£o (franquia)
        if 'belongs_to_collection' in movie_data and movie_data['belongs_to_collection']:
            collection_info = movie_data['belongs_to_collection']
            collection_id = collection_info['id']
            collection_name = collection_info['name']

            # Se a cole√ß√£o ainda n√£o foi processada, buscar os detalhes da cole√ß√£o
            if collection_id not in collections_found:
                collection_detail_url = f"https://api.themoviedb.org/3/collection/{collection_id}?api_key={API_KEY}&language=pt-BR"
                collection_response = requests.get(collection_detail_url)
                collection_response.raise_for_status()
                collection_details = collection_response.json()

                # Armazenar os filmes da cole√ß√£o
                collection_movies = []
                for part in collection_details.get('parts', []):
                    collection_movies.append({
                        'id': part['id'],
                        'title': part['title'],
                        'release_date': part.get('release_date', '')
                    })
                collections_found[collection_id] = {
                    'name': collection_name,
                    'movies': collection_movies
                }

            # Adicionar o filme atual √† lista de filmes de franquia brasileiros,
            # com o nome da cole√ß√£o a qual pertence.
            brazilian_franchise_movies.append({
                'movie_id': movie_id,
                'movie_title': movie_title,
                'collection_id': collection_id,
                'collection_name': collection_name
            })
            print(f"  Encontrado: '{movie_title}' faz parte da cole√ß√£o '{collection_name}'")

        time.sleep(0.05) # Pequeno atraso para evitar sobrecarregar a API (50ms)

    except requests.exceptions.HTTPError as e:
        print(f"Erro HTTP para o filme ID {movie_id} ({movie_title}): {e}")
    except requests.exceptions.RequestException as e:
        print(f"Erro de requisi√ß√£o para o filme ID {movie_id} ({movie_title}): {e}")
    except Exception as e:
        print(f"Erro inesperado para o filme ID {movie_id} ({movie_title}): {e}")

print("\n--- Buscas por franquias conclu√≠das ---")

# Filtrar apenas as cole√ß√µes que cont√™m filmes brasileiros encontrados
relevant_collections = {}
for movie_info in brazilian_franchise_movies:
    col_id = movie_info['collection_id']
    if col_id not in relevant_collections:
        relevant_collections[col_id] = collections_found[col_id]

if not relevant_collections:
    print("Nenhum filme brasileiro encontrado pertence a uma cole√ß√£o (franquia) no TMDB.")
else:
    print("\n--- Franquias Brasileiras Encontradas ---")
    for col_id, col_data in relevant_collections.items():
        print(f"\nCole√ß√£o: {col_data['name']} (ID: {col_id})")
        print("  Filmes na cole√ß√£o (com IDs do TMDB):")
        # Filtrar os filmes da cole√ß√£o que est√£o no seu df_brasileiros
        movies_in_collection_from_df = [
            f"    - {m['title']} (ID: {m['id']})"
            for m in col_data['movies']
            if m['id'] in df_brasileiros['id'].tolist()
        ]
        if movies_in_collection_from_df:
            for m_str in movies_in_collection_from_df:
                print(m_str)
        else:
            print("    (Nenhum filme desta cole√ß√£o encontrado no seu DataFrame brasileiro filtrado)")

    print("\n--- Resumo dos filmes brasileiros que fazem parte de franquias ---")
    if brazilian_franchise_movies:
        for movie in brazilian_franchise_movies:
            print(f"- '{movie['movie_title']}' (ID: {movie['movie_id']}) pertence √† cole√ß√£o '{movie['collection_name']}' (ID: {movie['collection_id']})")
    else:
        print("Nenhum filme brasileiro encontrado pertence a uma franquia.")
```

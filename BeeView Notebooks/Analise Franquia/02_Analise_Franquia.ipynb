# 🎬 Projeto de Análise de Filmes Sequenciais: Vale a Pena Produzir Continuações?

**Uma análise aprofundada sobre os fatores que determinam o sucesso de franquias cinematográficas, com foco em métricas quantitativas e aspectos qualitativos cruciais.**

---

## 1. Introdução e Justificativa

O presente projeto visa responder à pergunta central da indústria cinematográfica: **Vale a pena fazer uma sequência?** Através da análise de um vasto conjunto de dados de filmes, incluindo o "TMDB Movies Dataset 2023" do Kaggle e informações atualizadas via API do TMDB, este estudo se propõe a identificar padrões e insights que possam orientar decisões de produção.

A crescente tendência de investir em sequências e universos compartilhados, em um mercado com um volume expressivo de lançamentos anuais, justifica a relevância desta análise.

---

## 2. Objetivos do Projeto

### 2.1. Objetivo Principal
* Responder à pergunta: **Vale a pena fazer uma sequência cinematográfica?**

### 2.2. Objetivos Secundários
A investigação buscará identificar:
* Gêneros mais promissores para o desenvolvimento de franquias de sucesso.
* A influência da avaliação do público no desempenho comercial e crítico das continuações.
* O impacto do intervalo de tempo entre os lançamentos no sucesso das sequências.

---

## 3. Dados e Métricas

### 3.1. Fontes de Dados
* Kaggle: "TMDB Movies Dataset 2023"
* API do TMDB (para informações atualizadas)

### 3.2. Métricas Quantitativas Chave
A investigação se concentrará em:
* Bilheteria (`revenue`)
* Investimento (`budget`)
* Retorno sobre Investimento (`ROI`)
* Avaliações do público (`vote_average`)
* Popularidade (`popularity`)
* Gêneros (`genres`)

### 3.3. Definição de Sucesso (para este projeto)
Um filme será caracterizado como um sucesso se atender aos seguintes critérios:
* ROI ≥ 1 (indicando lucratividade)
* Avaliação média (`vote_average`) ≥ 7,0
* Popularidade estável ou crescente dentro da franquia

---

## 4. Análise Qualitativa (Diferencial)

Um diferencial desta análise será a inclusão de aspectos qualitativos complexos, como:

* **Conclusão narrativa:** Se a história de um filme encontra um fechamento natural, a necessidade e a receptividade de uma continuação podem ser questionadas.
* **Mudanças em pessoal chave:** A morte de um ator principal ou a saída do diretor original são fatores que podem influenciar drasticamente o rumo e a aceitação de uma sequência.

---

## 5. Primeiras Observações e Direcionamentos da Análise

Pesquisas iniciais já apontam para tendências relevantes que serão exploradas com os dados:

* **Desempenho de Sequências vs. Originais:**
    * Estudos indicam que, em média, os filmes originais tendem a ter uma bilheteria e avaliação ligeiramente superiores às suas sequências diretas.
    * Contudo, sequências ainda podem ser altamente lucrativas e, em alguns casos, superar os originais, especialmente em franquias bem estabelecidas.

* **A Relevância da Avaliação do Público:**
    * A recepção crítica e a avaliação do público do filme original têm um impacto direto no interesse e no desempenho inicial de uma sequência.
    * Uma avaliação negativa pode comprometer significativamente o potencial de uma continuação.

* **O Fator Tempo:**
    * O intervalo entre os lançamentos é crucial. Um período ótimo parece situar-se entre 2 a 5 anos.
    * Intervalos muito curtos podem saturar o público, enquanto esperas muito longas podem diminuir o interesse e o "momentum" da franquia.

* **Gêneros com Maior Propensão a Franquias de Sucesso:**
    * Ação (especialmente filmes de super-heróis), Animação e Fantasia/Ficção Científica demonstram consistentemente um alto potencial para a criação de franquias duradouras e lucrativas.

* **Impacto de Mudanças Drásticas:**
    * **Morte de Atores Principais:** Pode levar a reescritas significativas, reescalações (frequentemente com resistência dos fãs), uso de CGI para recriar o ator ou a aposentadoria do personagem. O legado do ator e do personagem torna-se um ponto sensível para a continuação.
    * **Saída de Diretores:** É geralmente vista como um risco, podendo levar a uma queda na qualidade ou a uma mudança de tom que desagrade os fãs. No entanto, novos diretores também podem revitalizar uma franquia com novas perspectivas. A coesão da visão criativa é um desafio.

* **A Armadilha da Conclusão Narrativa:**
    * Continuar uma história que teve um desfecho satisfatório e completo exige uma justificativa narrativa muito forte.
    * Sequências que parecem forçadas ou que desfazem o impacto emocional do final anterior tendem a ser mal recebidas.

#### 3.4 CRIANDO A TABELA ROI PARA A ANALISE FUTURA


```python
import numpy as np
import pandas as pd

# Certifique-se de que tmdb_dataset_filtrado está definido e com a coluna 'ROI' calculada.
# Se você não executou o bloco de código completo (carregamento + limpeza + cálculo de ROI)
# desde que iniciou ou resetou a sessão do Colab, por favor, execute-o primeiro.

# 1. Encontrar o valor máximo de ROI que NÃO é infinito
# Filtramos os valores infinitos para pegar o máximo entre os valores finitos.
max_finite_roi = tmdb_dataset_filtrado[tmdb_dataset_filtrado['ROI'] != np.inf]['ROI'].max()

# Definir um valor de substituição para 'inf'.
# Usaremos 1.5 vezes o maior ROI finito. Isso garante que ele continua sendo o valor mais alto,
# mas agora é um número finito. Se por algum motivo não houver ROI finito positivo, usamos um valor padrão alto.
if pd.isna(max_finite_roi) or max_finite_roi <= 0: # Caso não haja ROI finito positivo ou seja 0
    replacement_value_for_inf = 1e9 # 1 bilhão, um número arbitrariamente grande
else:
    replacement_value_for_inf = max_finite_roi * 1.5

# 2. Substituir os valores infinitos na coluna 'ROI'
# Substituímos tanto np.inf (infinito positivo) quanto -np.inf (infinito negativo, menos comum aqui)
tmdb_dataset_filtrado['ROI'].replace([np.inf, -np.inf], replacement_value_for_inf, inplace=True)

print(f"Valores 'inf' (e -inf) na coluna 'ROI' foram substituídos por: {replacement_value_for_inf:.2f}")

print("\n--- Estatísticas Descritivas para a coluna ROI após tratamento ---")
print(tmdb_dataset_filtrado['ROI'].describe())

print("\n--- Primeiras 5 linhas do DataFrame com o ROI tratado (observe o novo máximo se houver) ---")
print(tmdb_dataset_filtrado.head())
```

```python
import requests
import time # Para adicionar um pequeno atraso entre as requisições à API

# --- CONFIGURAÇÃO DA API TMDB ---
API_KEY = '1b5d8e773c2596486bd43142c6485e8c' # <<<< SUBSTITUA PELA SUA CHAVE REAL DO TMDB >>>>
BASE_URL = 'https://api.themoviedb.org/3/movie/'

# Dicionário para armazenar informações das coleções/franquias encontradas
# A chave será o ID da coleção, e o valor será um dicionário com nome e lista de filmes.
collections_found = {}
# Lista para armazenar informações dos filmes brasileiros que são sequências
brazilian_franchise_movies = []

print("Iniciando a busca por franquias para os filmes brasileiros...")
print(f"Processando {len(df_brasileiros)} filmes...")

# Iterar sobre cada filme no DataFrame de filmes brasileiros
for index, row in df_brasileiros.iterrows():
    movie_id = row['id']
    movie_title = row['title']

    # Construir a URL da requisição para detalhes do filme
    url = f"{BASE_URL}{movie_id}?api_key={API_KEY}&language=pt-BR" # pt-BR para obter nomes de coleções em português se disponíveis

    try:
        response = requests.get(url)
        response.raise_for_status() # Levanta um erro para códigos de status HTTP ruins (4xx ou 5xx)
        movie_data = response.json()

        # Verificar se o filme pertence a uma coleção (franquia)
        if 'belongs_to_collection' in movie_data and movie_data['belongs_to_collection']:
            collection_info = movie_data['belongs_to_collection']
            collection_id = collection_info['id']
            collection_name = collection_info['name']

            # Se a coleção ainda não foi processada, buscar os detalhes da coleção
            if collection_id not in collections_found:
                collection_detail_url = f"https://api.themoviedb.org/3/collection/{collection_id}?api_key={API_KEY}&language=pt-BR"
                collection_response = requests.get(collection_detail_url)
                collection_response.raise_for_status()
                collection_details = collection_response.json()

                # Armazenar os filmes da coleção
                collection_movies = []
                for part in collection_details.get('parts', []):
                    collection_movies.append({
                        'id': part['id'],
                        'title': part['title'],
                        'release_date': part.get('release_date', '')
                    })
                collections_found[collection_id] = {
                    'name': collection_name,
                    'movies': collection_movies
                }

            # Adicionar o filme atual à lista de filmes de franquia brasileiros,
            # com o nome da coleção a qual pertence.
            brazilian_franchise_movies.append({
                'movie_id': movie_id,
                'movie_title': movie_title,
                'collection_id': collection_id,
                'collection_name': collection_name
            })
            print(f"  Encontrado: '{movie_title}' faz parte da coleção '{collection_name}'")

        time.sleep(0.05) # Pequeno atraso para evitar sobrecarregar a API (50ms)

    except requests.exceptions.HTTPError as e:
        print(f"Erro HTTP para o filme ID {movie_id} ({movie_title}): {e}")
    except requests.exceptions.RequestException as e:
        print(f"Erro de requisição para o filme ID {movie_id} ({movie_title}): {e}")
    except Exception as e:
        print(f"Erro inesperado para o filme ID {movie_id} ({movie_title}): {e}")

print("\n--- Buscas por franquias concluídas ---")

# Filtrar apenas as coleções que contêm filmes brasileiros encontrados
relevant_collections = {}
for movie_info in brazilian_franchise_movies:
    col_id = movie_info['collection_id']
    if col_id not in relevant_collections:
        relevant_collections[col_id] = collections_found[col_id]

if not relevant_collections:
    print("Nenhum filme brasileiro encontrado pertence a uma coleção (franquia) no TMDB.")
else:
    print("\n--- Franquias Brasileiras Encontradas ---")
    for col_id, col_data in relevant_collections.items():
        print(f"\nColeção: {col_data['name']} (ID: {col_id})")
        print("  Filmes na coleção (com IDs do TMDB):")
        # Filtrar os filmes da coleção que estão no seu df_brasileiros
        movies_in_collection_from_df = [
            f"    - {m['title']} (ID: {m['id']})"
            for m in col_data['movies']
            if m['id'] in df_brasileiros['id'].tolist()
        ]
        if movies_in_collection_from_df:
            for m_str in movies_in_collection_from_df:
                print(m_str)
        else:
            print("    (Nenhum filme desta coleção encontrado no seu DataFrame brasileiro filtrado)")

    print("\n--- Resumo dos filmes brasileiros que fazem parte de franquias ---")
    if brazilian_franchise_movies:
        for movie in brazilian_franchise_movies:
            print(f"- '{movie['movie_title']}' (ID: {movie['movie_id']}) pertence à coleção '{movie['collection_name']}' (ID: {movie['collection_id']})")
    else:
        print("Nenhum filme brasileiro encontrado pertence a uma franquia.")
```
